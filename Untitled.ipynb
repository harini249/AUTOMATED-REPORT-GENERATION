```python
df = pd.read_csv('test.csv')
df
```

```python
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Platform'] = label_encoder.fit_transform(df['Platform'])
df['Daily_Usage_Time (minutes)'] = label_encoder.fit_transform(df['Daily_Usage_Time (minutes)'])
df['Dominant_Emotion'] = label_encoder.fit_transform(df['Dominant_Emotion'])
df
```

```python
df.info()
```

```python
null_counts = df.isnull().sum()
null_counts
```

```python
# Fill missing values for numeric columns
df['User_ID'] = df['User_ID'].fillna(df['User_ID'].mean())  # Fill User_ID with mean (if numeric, otherwise leave)
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])  # Fill Age with mode (as it's object type)
df['Posts_Per_Day'] = df['Posts_Per_Day'].fillna(df['Posts_Per_Day'].mean())  # Fill with mean
df['Likes_Received_Per_Day'] = df['Likes_Received_Per_Day'].fillna(df['Likes_Received_Per_Day'].mean())  # Fill with mean
df['Comments_Received_Per_Day'] = df['Comments_Received_Per_Day'].fillna(df['Comments_Received_Per_Day'].mean())  # Fill with mean
df['Messages_Sent_Per_Day'] = df['Messages_Sent_Per_Day'].fillna(df['Messages_Sent_Per_Day'].mean())  # Fill with mean
```

```python
# Fill missing values for categorical columns
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])  # Fill Gender with mode
df['Platform'] = df['Platform'].fillna(df['Platform'].mode()[0])  # Fill Platform with mode
df['Dominant_Emotion'] = df['Dominant_Emotion'].fillna(df['Dominant_Emotion'].mode()[0])  # Fill with mode
```

```python
print(df.isnull().sum())
```

```python
# Descriptive statistics
print(df.describe())

# Group by dominant emotion and calculate average usage time
print(df.groupby('Dominant_Emotion')['Daily_Usage_Time (minutes)'].mean())

# Correlation between daily usage and posts per day
print(df[['Daily_Usage_Time (minutes)', 'Posts_Per_Day']].corr())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
sns.scatterplot(x='Daily_Usage_Time (minutes)', y='Posts_Per_Day', data=df)
plt.title('Social Media Usage vs Posts Per Day')
plt.xlabel('Daily Usage Time (minutes)')
plt.ylabel('Posts Per Day')
plt.show()

# Histogram
df['Daily_Usage_Time (minutes)'].hist(bins=20)
plt.title('Distribution of Daily Usage Time')
plt.xlabel('Daily Usage Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```
